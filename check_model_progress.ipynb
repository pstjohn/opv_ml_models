```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context='talk', style='ticks',
        color_codes=True, rc={'legend.frameon': False})

%matplotlib inline
```

```python
train = pd.read_csv('data/mol_train.csv')
valid = pd.read_csv('data/mol_valid.csv')
test = pd.read_csv('data/mol_test.csv')
```

```python
props = ['gap', 'homo', 'lumo', 'spectral_overlap', 'homo_extrapolated',
         'lumo_extrapolated', 'gap_extrapolated', 'optical_lumo_extrapolated']

runs = ['b3lyp_2D_mae_multitarget'] + ['b3lyp_2d_{}'.format(prop) for prop in props]

fig, ax = plt.subplots(nrows=1, ncols=2, sharex=True, sharey=True, figsize=(8,4))

ax[0].set_yscale('log')
ax[1].set_yscale('log')

for run in runs:
    
    log = pd.read_csv('{}/log.csv'.format(run), index_col=0)
    
    ax[0].plot(log.loss)
    ax[1].plot(log.val_loss, label=run.replace('edgenetwork', '3D'))
    
ax[1].legend(loc='center left', bbox_to_anchor=(1., .5))
sns.despine()

ax[0].set_xlabel('Epoch')
ax[1].set_xlabel('Epoch')

ax[0].set_ylabel('Loss (MAE)')

ax[0].set_title('Training')
ax[1].set_title('Validation')
```
